```python
df = pd.read_csv('evaluation_results.csv', dtype=str)
```

```python
def circled_to_int(char):
    circled_numbers = [
        '①', '②', '③', '④', '⑤', 
        '⑥', '⑦', '⑧', '⑨', '⑩',
        '⑪', '⑫', '⑬', '⑭', '⑮',
        '⑯', '⑰', '⑱', '⑲', '⑳'
    ]
    if char in circled_numbers:
    # Unicode code point of '①' is 0x2460
        return ord(char) - ord('①') + 1
    else:
        return char
```

```python
total = 0
correct = 0
for index, row in df.iterrows():
    model_answer = "".join([str(circled_to_int(i)) for i in str(row['model_answer'])])
    total += 1
    if str(model_answer) == str(row['correct_answer']):
        correct += 1
    # print(f"{row['file']} #{row['number']}: \t", f"{model_answer}" , row['correct_answer'])
```

```python
print(f"Accuracy: {correct/total} ({correct}/{total})")
```

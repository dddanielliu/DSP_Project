```python
from pypdf import PdfReader
import re

reader = PdfReader("22000_-職業安全管理學科(甲安).pdf")
number_of_pages = len(reader.pages)
all_text = []
for i in range(number_of_pages):
    page = reader.pages[i]
    text = page.extract_text()
    text = re.sub(r'Page\s*\d+\s+of\s+\d+', '', text, flags=re.IGNORECASE)
    # 再把多餘空白整理掉
    text = re.sub(r'[ \t]{2,}', ' ', text)
    text = re.sub(r'\n{2,}', '\n\n', text)
    text = text.strip()
    all_text.append(text)
    
# page = reader.pages[0]
# text = page.extract_text()
```

```python
import re
import pandas as pd

# 讀取 output.csv
with open('output.csv', 'r', encoding='utf-8') as f:
    content = f.read()

# 正則解析
pattern = re.compile(r'(\d+)\.\s*\((\d+)\)\s*(.+?)(?=\d+\.\s*\(|$)', re.DOTALL) 
results = []
for match in pattern.finditer(content):
    number = match.group(1)
    answer = match.group(2)
    question = match.group(3).replace('\n', ' ').strip()
    results.append({'number': number, 'answer': answer, 'question': question})

df = pd.DataFrame(results)
df.to_csv('parsed_questions.csv', index=False, encoding='utf-8-sig')
df.head()
```
